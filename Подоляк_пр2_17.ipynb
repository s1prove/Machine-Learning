```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
data = pd.read_html(url, header=0)[2]

#Task1: Виведіть перші 5 рядків таблиці. 

df_1 = data.head(5)
print(df_1)
data
```

```python
#Task2: Визначте кількість рядків і стовпчиків за допомогою атрибуту shape 

num_of_rows, num_of_columns = data.shape
print(f"Rows = {num_of_rows}, Columns = {num_of_columns}")
```

```python
#Task3: Визначте загальну кількість років та кількість місяців у даних. 

num_of_years = len(data["Рік"])
num_of_months = num_of_years * 12
print(f"Number of years = {num_of_years}, number of months = {num_of_months}")
```

```python
#Task4: Замінити пропущені значення  «-» в таблиці на NaN 

data_clean = data.replace(to_replace = "—", value = np.NaN)
data_clean
```

```python
#Task5: Визначити тип стовпців за допомогою dtypes. 

data.dtypes
```

```python
#Task6: Видаліть передостанній і останній стовпчики 
#( Кількість кгхліба[20] на середнюзарплату,  Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею.

new_data = data_clean.drop(["Кількість кг хліба[20] на середню зарплату"], axis = 1)
new_data1 = new_data.drop(["Кількість БігМаків на середню зарплату"], axis = 1)
print(new_data1)
```

```python
#Task7: Замінити тип нечислових даних на числові.

for col in new_data1.columns[1:]:
    new_data1[col] = new_data1[col].str.replace(',', '.').str.extract('(\d+.\d+)').astype(float)
print(new_data1.dtypes)
print(new_data1)
```

```python
#Task8: Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len) 

new_data1.iloc[:, 1:].isnull().sum()/len(new_data1)
```

```python
#Task9: Замініть всі пропущені значення на середні значення по (стовпцю). 

from numpy.lib.function_base import average
x = new_data1.mean(axis=1)
for i, col in enumerate(new_data1):
    new_data1.iloc[:, i] = new_data1.iloc[:, i].fillna(x)
new_data1
```

```python
#Task10: Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату. 

max_value = new_data1['В національній валюті'].max()  
max_row = new_data1.loc[new_data1['В національній валюті'].idxmax()]['Рік'] 
print(f"The biggest salary: {max_value}")
print(f"Year: {max_row}")
```

```python
#Task11: Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату.

max_value = new_data1['В доларах США'].min() 
max_row = new_data1.loc[new_data1['В доларах США'].idxmin()]['Рік']  
print(f"The smallest salary: {max_value}")
print(f"Year: {max_row}")
```

```python
#Task12: Виведіть роки, коли зарплата в грн була вище середньої. 

average_salary = new_data1['В національній валюті'].mean()
above_average_years = new_data1.loc[new_data1['В національній валюті'] > average_salary]['Рік']

print("\nYears with sallary bigger than average:")
print(above_average_years.to_list())
```

```python
#Task13: Порівняйте  зарплату  в грн за останні два доступних роки

value1 = new_data1.at[30, 'В національній валюті']
value2 = new_data1.at[31, 'В національній валюті']

if value2 > value1:
        result = f"Зарплата в 2022 році ({value2} грн) вища за зарплату в 2021 році ({value1} грн)."
elif value2 < value1:
    result = f"Зарплата в 2022 році ({value2} грн) нижча за зарплату в 2021 році ({value1} грн)."
else:
    result = f"Зарплата в 2022 році ({value2} грн) дорівнює зарплаті в 2021 році ({value1} грн)."

print(result)
```

```python
#Task14: Створіть графік, що відображає  зарплату в Україні з часом (за роками).

x = new_data1.plot(x='Рік', y='В національній валюті', title='Зарплата в Україні за роками', figsize=(10,6), grid=True)
x.set_ylabel("мільйони грн.")
```

```python
#Task15: Видаліть останній рік з таблиці та виведіть оновлену таблицю. 

new_data1.drop([31], inplace=True)
new_data1
```

```python
#Task16: Побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року. 


years = new_data1['Рік'][:-1] 
salaries = new_data1['В національній валюті'][:-1]


plt.figure(figsize=(20, 6))
plt.plot(years, salaries, marker='o', color='b')
plt.title('Середня зарплата в Україні за роки (без останнього)')
plt.xlabel('Рік')
plt.ylabel('Середня зарплата (грн)')
plt.grid(True)

plt.show()
```

```python
#Task17: Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. 
#Вкажіть, чи є різниця, і в якому напрямку. 

last_five_years = new_data1.tail(5)

last_two_years = new_data1.tail(2)

salary_for_5 = last_five_years['В національній валюті'].sum()

salary_for_2 = last_two_years['В національній валюті'].sum()

if salary_for_2 > salary_for_5:
    result = "Зарплата за останні два роки вища за зарплату за останні 5 років."
elif salary_for_2 < salary_for_5:
    result = "Зарплата за останні два роки нижча за зарплату за останні 5 років."
else:
    result = "Зарплата за останні два роки дорівнює зарплаті за останні 5 років."

print(result)
```
